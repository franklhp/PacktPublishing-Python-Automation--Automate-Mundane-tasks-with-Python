# Web Scraping -  Mini project

### Problem

* Count total number of views for all video on 1st page of youtube for given keyword 
* Keyword : Python

### Solution

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
res = requests.get('https://www.youtube.com/results?search_query=python')
```

```python
res.text
```

```python
<div class="yt-lockup-meta "><ul class="yt-lockup-meta-info"><li>3 months ago</li><li>205,655 views</li></ul></div>
```

```python
bs = BeautifulSoup(res.text, 'lxml')
```

```python
elements = bs.find_all('ul', class_='yt-lockup-meta-info')
```

```python
len(elements)
```

```python
totalView = 0
for ele in elements:
    lis = ele.findChildren()
    for li in lis:
        if li.string.endswith('views'):
            temp = li.text
            temp = temp.replace(' views','')
            temp = temp.replace(',','')
            totalView = totalView + int(temp)
print(totalView)
```

```python
total = 0
for ele in elemts:
    lis = ele.findChildren()
    for li in lis:
        if li.string.endswith('views'):
            temp = li.string
            temp = temp.replace(' views','')
            temp = temp.replace(',','')
            total = total + int(temp)
            print(temp)
print('Total : '+ str(total))
```
